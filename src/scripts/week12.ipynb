```python
import csv
```

```python
f = open('../../기온 공공 데이터.csv', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)
```

```python
first_data = next(data)

records = {
    "low_temp": (float(first_data[3]), first_data[0]),
    "high_temp": (float(first_data[4]), first_data[0]),
    "high_diff": (float(first_data[4]) - float(first_data[3]), first_data[0])
}

for row in data:
    day, _, _, low, high = row
    low, high = float(low), float(high)
    diff = high - low

    if diff > records["high_diff"][0]:
        records["high_diff"] = (diff, day)
    if low < records["low_temp"][0]:
        records["low_temp"] = (low, day)
    if high > records["high_temp"][0]:
        records["high_temp"] = (high, day)

f.close()
```

```python
print(f"일교차가 가장 큰 날: {records['high_diff'][0]}\n날짜: {records['high_diff'][1]}\n")
print(f"최저기온이 가장 낮은 날: {records['low_temp'][0]}\n날짜: {records['low_temp'][1]}\n")
print(f"최고기온이 가장 높은 날: {records['high_temp'][0]}\n날짜: {records['high_temp'][1]}")
```

```python
f = open('../../지하철 노선별 역별 이용현황.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
```

```python
k = 3
top_k, bot_k = [], []

for _ in range(k):
    row = next(data)
    station, pass_num = row[3], int(row[4].replace(',', "")) + int(row[5].replace(',', ""))
    top_k.append((station, pass_num))
    bot_k.append((station, pass_num))

top_k = sorted(top_k, key=lambda tup: tup[1], reverse=True)
bot_k = sorted(bot_k, key=lambda tup: tup[1])
```

```python
for row in data:
    station, pass_num = row[3], int(row[4].replace(',', "")) + int(row[5].replace(',', ""))

    if any(pass_num > item[1] for item in top_k):
        top_k[2] = (station, pass_num)
        top_k = sorted(top_k, key=lambda tup: tup[1], reverse=True)
    
    if any(pass_num < item[1] for item in bot_k):
        bot_k[2] = (station, pass_num)
        bot_k = sorted(bot_k, key=lambda tup: tup[1])

f.close()
```

```python
print("이용객이 가장 많은 역 순위")
for idx, top in enumerate(top_k):
    print(f"{idx + 1}. {top[0]} ({top[1]}명)")

print("\n이용객이 가장 적은 역 순위")
for idx, bot in enumerate(bot_k):
    print(f"{idx + 1}. {bot[0]} ({bot[1]}명)")
```
